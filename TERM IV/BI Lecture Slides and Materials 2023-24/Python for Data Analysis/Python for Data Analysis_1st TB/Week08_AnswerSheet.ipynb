```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Week06_listings.csv
df = pd.read_csv('Week06_listings.csv')
```

```python
# 1- Draw the Scatter-plot of “latitude” and “longitude” values of houses. 
# Filter the houses in which select only the houses with price 300.
df2 = df[df.price==300]
# Solution 1
sns.regplot('latitude', 'longitude', data=df2)
# Solution 2
df2.plot(kind = 'scatter', x = 'latitude', y = 'longitude')
```

```python
# 2- Plot line chart for “latitude” and “longitude” columns 
# and compare the line chart result with values of the “latitude” and “longitude” columns in the dataset.

fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
plt.plot(df['latitude'])
ax2 = fig.add_subplot(2,1,2)
plt.plot(df['longitude'])
```

```python
# 3- Plot a bar chart for the frequency of the “room_type” in the dataset.

df.groupby('room_type')['id'].count().plot(kind = 'bar')
```

```python
# 4- Create a histogram chart for each numerical columns in dataset 
# (Remember subplot in histogram).

df.describe()
# The numerical columns are: 
# latitude 
# longitude 
# price 
# minimum_nights 
# number_of_reviews
# reviews_per_month
# calculated_host_listings_count
# availability_365

fig = plt.figure(figsize = (10, 10))
ax1 = fig.add_subplot(4,2,1)
df['latitude'].plot(kind = 'hist')

ax2 = fig.add_subplot(4,2,2)
df['longitude'].plot(kind = 'hist')

ax3 = fig.add_subplot(4,2,3)
df['price'].plot(kind = 'hist')

ax4 = fig.add_subplot(4,2,4)
df['minimum_nights'].plot(kind = 'hist')

ax5 = fig.add_subplot(4,2,5)
df['number_of_reviews'].plot(kind = 'hist')

ax6 = fig.add_subplot(4,2,6)
df['reviews_per_month'].plot(kind = 'hist')

ax7 = fig.add_subplot(4,2,7)
df['calculated_host_listings_count'].plot(kind = 'hist')

ax8 = fig.add_subplot(4,2,8)
df['availability_365'].plot(kind = 'hist')
```

```python
# 5- Create a boxplot for “price” in the dataset. 

df.boxplot(column = ['price'], figsize = (5, 10))
```
