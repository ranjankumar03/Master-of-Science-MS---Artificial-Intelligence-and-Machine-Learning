```python
import pandas as pd

df = pd.read_csv("Week06_listings.csv")
df
```

```python
# 1- Groupby dataset using column “room_type” and create statistics for “price”? 
# What are the minimum, maximum price, sum of price, mean of price for each room type?

grouped = df[["price","room_type"]].groupby("room_type")
grouped["price"].agg(['min','max','sum','mean'])
```

```python
# 2- Groupby dataset using column “room_type” and create statistics for “minimum_nights”?
# what are the minimum, maximum minimum nights for each room type?

grouped = df[["minimum_nights","room_type"]].groupby("room_type")
grouped.describe()
```

```python
# 3- Create a function that returns the max and min price difference and use this function in agg method.

def min_max_difference(arr):
    return arr.max()-arr.min()

grouped = df[["price","room_type"]].groupby("room_type")
grouped["price"].agg(['min','max',min_max_difference])
```

```python
# 4- Show the group names and their data for grouped data by “room_type”.

for name, group in df.groupby("room_type"):
    print(name)
    print(group)
```

```python
# 5- Find which “room_type” has most “calculated_host_listings_count”. 

grouped = df[["room_type","calculated_host_listings_count"]].groupby("room_type").agg(['max'])
print(grouped)
grouped.max()
```
